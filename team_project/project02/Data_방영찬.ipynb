# 1980~2020년 교통수단별 교통사고 현황

```python
# 모듈 로딩 --------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 파일 관련 변수 선언 ------------------------
DIR_PATH='./Data_영찬/'
FILE_NAME=DIR_PATH+'교통수단별 현황.xlsx'
```

```python
from matplotlib import font_manager as fm, rc
font_path='../Fonts/malgun.ttf'
font_name=fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
# (1) Excel FILE => DataFrame으로 로딩----------------
tpDF=pd.read_excel(FILE_NAME)
tpDF.info()
```

# (2) 데이터 정보 확인--------------------------------------------------------------------

```python
print('\nwt.columns->\n', tpDF.columns)
print('\nwt.head(10)->\n', tpDF.head(10))
print('\nwt->\n', tpDF)
```

## 결측치 갯수 파악 -------------

```python
print('\ntpDF.isnull()->\n', tpDF.isnull())
print('\ntpDF.isnull().sum()->\n', tpDF.isnull().sum())
```

## shape, ndim 정보 확인

```python
print('\ntpDF.shape->\n', tpDF.shape)
print('\ntpDF.ndim->\n', tpDF.ndim)
```

# 고유값, 값의개수

```python
for a in tpDF.columns:
    print('tpDF의 각칼럼의 고유값->\n', tpDF[a].unique())

tpDF_r_1=tpDF.iloc[:].replace('...', None)
tpDF_r_2=tpDF_r_1.iloc[:].replace('-', None)
tpDF_r_3=tpDF_r_2.iloc[:].replace(np.nan, None)

for a in tpDF.columns:
    print('\ntpDF_r의 각칼럼의 고유값->\n', tpDF_r_3[a].unique())

print('tpDF_r_3->\n', tpDF_r_3)
```

```python
tpDF_r_d=tpDF_r_3.drop(0)

print('tpDF_r_d->\n', tpDF_r_d)
```

# 1980 ~ 2020년까지의 각 발생건수와 사망자수 합 총합 ---------------------

```python
tpDF_r_d_s_1=tpDF_r_d.drop([1, 2])
tpDF_r_d_s_2=tpDF_r_d_s_1.drop('Unnamed: 0', axis=1)
print('tpDF_r_d_s_1->\n', tpDF_r_d_s_1)
print('tpDF_r_d_s_2->\n', tpDF_r_d_s_2)
```

    # 발생건수 합

```python
for a in range(5):
     print(f'(사건 수 합)tpDF_r_d_s_2["Unnamed: {2*a+1}"].sum()->\n',
           tpDF_r_d_s_2[f"Unnamed: {2*a+1}"].sum())
print('------------------------\n')
```

    #사망자수 합

```python
for a in range(5):
     print(f'(사망자 수 합)tpDF_r_d_s_2["Unnamed: {2*a+2}"].sum()->\n',
           tpDF_r_d_s_2[f"Unnamed: {2*a+2}"].sum())
print('------------------------\n')
```

   # --> 1980 ~ 2020 자동차의 발생건수와 사망자수 합이
   #       가장 많았음
   # --> 1980 ~ 2020 항공기의 발생건수와 사망자수 합이
   #       가장 적었음

   # <span style="color:red">*** 철도와 선박을 비교해봤을 때</span>
   # <span style="color:red">*** (철도의 발생건수 합 < 선박의 발생건수 합)이지만</span>
   # <span style="color:red">*** (철도의 사망자수 합 > 선박의 사망자수 합)을 알 수 있었습니다.</span>
   # 특이점으로는 항공기가 발생건수에 비해 사망자수가 많은 것을 알 수 있었습니다.

```python
# 인덱스 변경 ----------------------------
print('-----------')
tpDF_idx=tpDF_r_d_s_1.set_index('Unnamed: 0')
print('\ntpDF_idx->\n', tpDF_idx)
# -------------------------------
# 1980 ~ 2020년 자동차 사고 최대 발생건수와 사망자수 해당 년도
tpDF_z_1=tpDF.iloc[:].replace('...', 0)
tpDF_z_2=tpDF_z_1.iloc[:].replace('-', 0)
tpDF_z_3=tpDF_z_2.iloc[:].replace(np.nan, 0)
print('\ntpDF_z_3->\n', tpDF_z_3)
tpDF_z_d=tpDF_z_3.drop(0)
print('tpDF_z_d->\n', tpDF_z_d)
tpDF_z_d_m_1=tpDF_z_d.drop([1, 2])
tpDF_z_d_m_2=tpDF_z_d_m_1.drop('Unnamed: 0', axis=1)
print('tpDF_z_d_m_1->\n', tpDF_z_d_m_1)
print('tpDF_z_d_m_2->\n', tpDF_z_d_m_2)
tpDF_z_idx=tpDF_z_d_m_1.set_index('Unnamed: 0')
print('tpDF_z_idx->\n', tpDF_z_idx)
```

    # 최대 발생건수

```python
for a in range(5):
    print(f'(최대 발생건수)tpDF_z_idx["Unnamed: {2*a+1}"].max()->\n',
           tpDF_z_idx[f"Unnamed: {2*a+1}"].max())
    print('해당 년도-> ', tpDF_z_idx.index[np.argmax(tpDF_z_idx[f"Unnamed: {2*a+1}"])])
print('------------------------\n')
```

    #최대 사망자수

```python
for a in range(5):
    print(f'(최대 사망자수)tpDF_z_idx["Unnamed: {2*a+2}"].max()->\n',
           tpDF_z_idx[f"Unnamed: {2*a+2}"].max())
    print('해당 년도-> ', tpDF_z_idx.index[np.argmax(tpDF_z_idx[f"Unnamed: {2*a+2}"])])
print('------------------------\n')
```
