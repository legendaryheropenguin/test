```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
accident2020=pd.read_csv('./traffic_2020.csv',encoding='cp949',header=None,thousands = ',')
accident2019=pd.read_csv('./traffic_2019.csv',encoding='cp949',header=None,thousands = ',')
accident2018=pd.read_csv('./traffic_2018.csv',encoding='cp949',header=None,thousands = ',')
accident2017=pd.read_csv('./traffic_2017.csv',encoding='cp949',header=None,thousands = ',')
accident2016=pd.read_csv('./traffic_2016.csv',encoding='cp949',header=None,thousands = ',')
```

```python
accident2020
```

```python
first_list=[accident2020,accident2019,accident2018,accident2017,accident2016]
second_list=[]
```

```python
def data_func(file_name):
    file_name.drop([0,1,2],inplace=True)
    file_name=file_name[[0,1,7]].transpose()
    file_name.fillna("",inplace=True)
    file_name=file_name.rename(columns=file_name.iloc[0])
    file_name.drop(0,inplace=True)
    file_name=file_name.set_index(file_name.columns[0])
    return file_name
```

```python
for i in first_list:
    second_list.append(data_func(i))
```

```python
total_data=pd.concat([second_list[0],second_list[1],second_list[2],second_list[3],second_list[4]])
total_data.columns
```

```python
car_data=total_data.iloc[::2]
car_data['자동차등록대수']=[2020,2019,2018,2017,2016]
car_data=car_data.set_index('자동차등록대수')
car_data
```

```python
acc_data=total_data.iloc[1::2]
acc_data['사고건수']=[2020,2019,2018,2017,2016]
acc_data=acc_data.set_index('사고건수')
```

```python
for i in range(5):
    for j in range(18):
        accdata1 = acc_data.iloc[i,j].replace(",","")
        cardata1 = car_data.iloc[i,j].replace(",","")
        acc_data.iloc[i,j] = (int(accdata1)/int(cardata1))*100
acc_data['사고건수/자동차등록대수']=[2020,2019,2018,2017,2016]
final_acc=acc_data.set_index('사고건수/자동차등록대수')        
final_acc.columns=['Total','Seoul','Busan','Gyeonggi-do','Gangwon-do','Chungbuk','Chungnam','Jeonbuk','Jeonnam','Gyeongbuk','Gyeongnam','Jeju','Daegu','Incheon','Gwangju','Daejeon','Ulsan','Sejong']
final_acc
```

# 1번 꺽은선 그래프 : 지역별 5년간 자동차 사고확률 변화추이 비교

```python
# 그래프 사이즈 설정
plt.figure(figsize=(12, 13))

# 그래프 데이터 설정
#plt.plot(final_acc)
#plt.plot(final_acc)

# 각 선당 라벨링
for idx, col in enumerate(final_acc.columns):
    plt.plot(final_acc.index, final_acc[col])
    
    if idx%2:
            plt.annotate(col, xy=(2020, final_acc.iloc[0][idx]), xytext=(2020, final_acc.iloc[0][idx]), fontsize=14, ha='center',
            arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))
    else:
            plt.annotate(col, xy=(2019, final_acc.iloc[1][idx]), xytext=(2019, final_acc.iloc[0][idx]), fontsize=14, ha='center',
            arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))
    
# 그래프 이름, 축이름, 범례 설정
# plt.title('Car Accident Comparison Analysis' , fontsize=20) 
# plt.ylabel('accident / car count [%]', fontsize=20)  # y축 라벨
# plt.xlabel('year',fontsize=20)
# plt.legend(['Total','Seoul','Busan','Gyeonggi-do','Gangwon-do','Chungbuk','Chungnam','Jeonbuk','Jeonnam','Gyeongbuk','Gyeongnam','Jeju','Daegu','Incheon','Gwangju','Daejeon','Ulsan','Sejong'],title = "Region")   # 범례
plt.show()
```

# 2번 막대 그래프 : 지역별 5년간 자동차 사고확률의 평균값 비교

```python
# acc_avr = (사고건수/차 등록 개수)의 평균 계산 정렬하여 저장.

# 칼럼별(지역별) 5년간 사고확률 평균 계산
acc_avr=final_acc.mean()
acc_avr_sort =acc_avr.sort_values()
acc_avr_y =acc_avr_sort.index
# 데이터프레임으로 변경 후 정렬 (순위를 알기 위함.)
#df=pd.DataFrame(acc_avr)
#df=sorted(df[0])
```

```python
# 그래프 사이즈 
plt.figure(figsize=(15, 12))

# x축. y축 데이터 값 설정
plt.barh(acc_avr_y,acc_avr_sort)

# 그래프 이름, 축 이름 설정
plt.title('Car Accident Comparison Analysis : Average of traffic accidents', fontsize=20) 
plt.ylabel('Region', fontsize=20)  
plt.xlabel('Average of traffic accidents by region for 5 years',fontsize=20)

plt.show()
```
